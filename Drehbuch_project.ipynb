```python
%%writefile /content/CampusTV-Drehbuch-Tool-paket2/db_helper.py
import psycopg2
import pandas as pd
from sqlalchemy import create_engine, text

SUPABASE_HOST     = "aws-0-eu-west-1.pooler.supabase.com"
SUPABASE_PORT     = "6543"
SUPABASE_DATABASE = "postgres"
SUPABASE_USER     = "postgres.fdfislabvftrinoraaby"
SUPABASE_PASSWORD = "55640363Amen"

def get_engine(host=SUPABASE_HOST, port=SUPABASE_PORT, database=SUPABASE_DATABASE, user=SUPABASE_USER, password=SUPABASE_PASSWORD):
    url = f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}?sslmode=require"
    try:
        engine = create_engine(url)
        with engine.connect() as c:
            c.execute(text("SELECT 1"))
        print("✅ Erfolgreich mit Supabase verbunden!")
        return engine
    except Exception as e:
        print(f"❌ Verbindung fehlgeschlagen: {e}")
        return None

def create_sandbox_table(engine):
    if engine is None:
        return
    q = """CREATE TABLE IF NOT EXISTS szenen (
        id SERIAL PRIMARY KEY,
        szene_nummer INT NOT NULL,
        titel VARCHAR(255),
        ort VARCHAR(100),
        zeit VARCHAR(50),
        inhalt TEXT
    );"""
    with engine.connect() as conn:
        conn.execute(text(q))
        conn.commit()
    print("🎭 Tabelle 'szenen' steht bereit!")

def insert_test_data(engine):
    if engine is None:
        return
    with engine.connect() as conn:
        count = conn.execute(text("SELECT COUNT(*) FROM szenen")).scalar()
        if count == 0:
            conn.execute(text("""
                INSERT INTO szenen (szene_nummer, titel, ort, zeit, inhalt) VALUES
                (1, 'Das Treffen im Studio', 'CAMPUS TV STUDIO', 'TAG', 'Mahmoud testet die Datenbankverbindung.'),
                (2, 'Der Durchbruch', 'BHT LABOR', 'NACHT', 'Der Code läuft!');
            """))
            conn.commit()
            print("✍️ Test-Daten eingefügt!")
        else:
            print(f"ℹ️ Bereits {count} Einträge vorhanden.")

def fetch_all_scenes(engine):
    if engine is None:
        return None
    return pd.read_sql_query("SELECT * FROM szenen ORDER BY szene_nummer;", engine)
```
