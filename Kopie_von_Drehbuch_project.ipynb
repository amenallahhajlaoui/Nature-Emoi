```python
#letzte zahler
import re
import torch
import transformers
from weasyprint import HTML

# 1. Pfad definieren und Modell direkt aus dem Drive laden
drive_model_path = "/content/drive/MyDrive/CampusTV_AI_Model"
print("⚡ Lade Modell direkt aus dem Google Drive auf die GPU...")

pipeline = transformers.pipeline(
    "text-generation",
    model=drive_model_path,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto"
)
print("✅ Modell erfolgreich aus dem Drive geladen und startklar!\n")

# 2. Drehbuch-Szene für die Analyse
beispiel_szene = """
SZENE 3. CAMPUS TV STUDIO - TAG

MAHMOUD (20er, konzentriert) sitzt vor einem Monitor. Er tippt auf der Tastatur.
Neben ihm steht PROF. DR. STRZEBKOWSKI (50er, ruhig) mit verschränkten Armen.

MAHMOUD
Der Datenbank-Sync steht, Professor. Die Daten fließen live.

PROF. STRZEBKOWSKI
Gut. Aber läuft das System auch stabil, wenn wir das Sprachmodell dranhängen?
"""

prompt = f"""
Du bist ein KI-Assistent für ein Drehbuch-Tool. Analysiere diese Szene kurz und knackig:
1. Zusammenfassung: Was passiert in 1-2 Sätzen?
2. Kamera-Vorschlag: Welche Kameraeinstellung passt am besten zu Beginn der Szene?

Szene:
{beispiel_szene}
"""

# 3. KI-Inferenz ausführen
print("🤖 KI analysiert die Szene...")
messages = [{"role": "user", "content": prompt}]
outputs = pipeline(messages, max_new_tokens=200)
ki_text = outputs[0]["generated_text"][-1]["content"]

print("\n--- KI ANALYSE ERGEBNIS IM ARBEITSSPEICHER ---")
print(ki_text)

# 4. Text dynamisch per Regex für das PDF aufteilen
zusammenfassung_match = re.search(r"1\.\s*(.*?)(?=2\.)", ki_text, re.DOTALL)
kamera_match = re.search(r"2\.\s*(.*)", ki_text, re.DOTALL)

zusammenfassung_live = zusammenfassung_match.group(1).strip() if zusammenfassung_match else ki_text
kamera_live = kamera_match.group(1).strip() if kamera_match else "Siehe oben."

# 5. Elegantes HTML/CSS-Layout befüllen
html_content = f"""
<!DOCTYPE html>
<html>
<head>
<meta charset="utf-8">
<style>
    @page {{
        size: A4;
        margin: 20mm 15mm;
        @bottom-right {{
            content: "Seite " counter(page);
            font-family: 'Arial', sans-serif;
            font-size: 9pt;
            color: #7f8c8d;
        }}
    }}
    body {{
        font-family: 'Arial', sans-serif;
        color: #2c3e50;
        line-height: 1.6;
    }}
    .header {{
        background-color: #1a252f;
        color: #ffffff;
        padding: 25px;
        border-radius: 6px;
        margin-bottom: 30px;
    }}
    .header h1 {{ margin: 0; font-size: 20pt; }}
    .header .meta {{ font-size: 10pt; color: #3498db; font-weight: bold; margin-top: 5px; }}
    h2 {{ color: #1a252f; font-size: 13pt; border-bottom: 2px solid #34495e; padding-bottom: 5px; margin-top: 25px; }}
    .live-box {{ background-color: #f8f9fa; border-left: 4px solid #3498db; padding: 15px; margin: 15px 0; }}
</style>
</head>
<body>
    <div class="header">
        <h1>CampusTV-Drehbuch-Tool</h1>
        <div class="meta">Generierte Live-Analyse aus Qwen-2.5</div>
    </div>

    <h2>1. Automatische Zusammenfassung</h2>
    <div class="live-box">{zusammenfassung_live}</div>

    <h2>2. Kamera- & Regievorschlag</h2>
    <div class="live-box" style="border-left-color: #2ecc71;">{kamera_live}</div>
</body>
</html>
"""

# 6. Als fertiges PDF ausgeben
pdf_filename = "CampusTV_Live_Analyse.pdf"
print("\n📄 Erstelle PDF-Datei direkt aus den KI-Ergebnissen...")
HTML(string=html_content).write_pdf(pdf_filename)
print(f"✅ Fertig! PDF wurde im Verzeichnis gespeichert: '{pdf_filename}'")
```
